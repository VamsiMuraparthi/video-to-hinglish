# Splitting the video into pieces of 4 min each for better translation

```python
from moviepy.video.io.VideoFileClip import VideoFileClip
```

```python
def split_video(input_video, output_prefix, duration):
    clip = VideoFileClip(input_video)
    total_duration = clip.duration
    num_parts = int(total_duration / duration) + 1
    
    for i in range(num_parts):
        start_time = i * duration
        end_time = min((i + 1) * duration, total_duration)
        subclip = clip.subclip(start_time, end_time)
        
        output_filename = f"{output_prefix}_part{i+1}.mp4"
        subclip.write_videofile(output_filename, codec='libx264')
        
    clip.reader.close()
    clip.audio.reader.close_proc()

if __name__ == "__main__":
    input_video = "input.mp4"  # Replace with your input video file
    output_prefix = "output_video"  # Prefix for output file names
    duration = 240  # Duration of each part in seconds (1 minute)
    
    split_video(input_video, output_prefix, duration)
```
